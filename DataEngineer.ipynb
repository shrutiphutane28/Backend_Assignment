Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

# **Exploratory Data Analysis**

**Data Cleaning**

```python
df.isnull().sum()
```

Fill missing REPORT_NAME with a placeholder

```python
df['REPORT_NAME'].fillna('Unknown Report', inplace=True)
```

Remove duplicates

```python
df = df.drop_duplicates()
```

Convert date columns to datetime

```python
df['CREATED_DATE_TIME'] = pd.to_datetime(df['CREATED_DATE_TIME'], errors='coerce')
```

```python
df['EXTRACTED_AT'] = pd.to_datetime(df['EXTRACTED_AT'], errors='coerce')
```

Convert REPORT_TYPE to category

```python
df['REPORT_TYPE'] = df['REPORT_TYPE'].astype('category')
```

```python
df
```

Handle USERS column with JSON

```python
def extract_display_name(json_str):
    try:
        parsed_json = json.loads(json_str)
        return parsed_json[0]['displayName'] if parsed_json else None
    except (json.JSONDecodeError, TypeError):
        return None

df['USERS'] = df['USERS'].apply(extract_display_name)
```

Convert column names to lowercase and replace spaces with underscores

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

Add 'unknown' category to categorical columns

```python
for col in df.select_dtypes(include='category').columns:
    df[col] = df[col].cat.add_categories('unknown')

df = df.fillna('unknown')
```

```python
df
```

```python
df.dtypes
```

Count the frequency of each report type

```python
report_type_counts = df['report_type'].value_counts()
print("Report Type Distribution:")
print(report_type_counts)
```

Plotting Report Type Distribution

```python
plt.figure(figsize=(10, 6))
report_type_counts.plot(kind='bar', color='skyblue')
plt.title("Report Type Distribution", fontsize=16)
plt.xlabel("Report Type", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Report Type Distribution Across Workspaces
workspace_report_type_distribution = df.groupby('workspace_name')['report_type'].value_counts().unstack()

# Plotting the Report Type Distribution Across Workspaces (Stacked Bar Chart)
workspace_report_type_distribution.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab20')
plt.title("Report Type Distribution Across Workspaces", fontsize=16)
plt.xlabel("Workspace Name", fontsize=12)
plt.ylabel("Number of Reports", fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Report Type Distribution Across Users (Heatmap)
user_report_type_distribution = df.groupby('modified_by')['report_type'].value_counts().unstack().fillna(0)
plt.figure(figsize=(12, 8))
sns.heatmap(user_report_type_distribution, annot=True, cmap='YlGnBu', fmt='g', cbar_kws={'label': 'Number of Reports'})
plt.title("Report Type Distribution Across Users (Heatmap)", fontsize=16)
plt.xlabel("Report Type", fontsize=12)
plt.ylabel("User", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Domain Report Distribution
domain_report_distribution = df['domain_id'].value_counts()

# Plotting Domain Report Distribution (Bar Chart)
plt.figure(figsize=(12, 6))
domain_report_distribution.plot(kind='bar', color='skyblue')
plt.title("Report Distribution per Domain", fontsize=16)
plt.xlabel("Domain", fontsize=12)
plt.ylabel("Number of Reports", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Ensure the 'created_date_time' column is in datetime format
df['created_date_time'] = pd.to_datetime(df['created_date_time'], errors='coerce')

# Temporal Trends - Reports Over Time (Created Date)
df['month_created'] = df['created_date_time'].dt.to_period('M')

# Count the number of reports created per month
monthly_report_creation = df['month_created'].value_counts().sort_index()

# Plotting Reports Over Time (Line Plot)
plt.figure(figsize=(12, 6))
monthly_report_creation.plot(kind='line', marker='o', color='lightcoral')
plt.title("Reports Created Over Time", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Reports", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# User Activity - Frequency of Report Modifications by Each User
user_activity = df['modified_by'].value_counts()

# Plotting User Activity (Bar Chart)
plt.figure(figsize=(12, 6))
user_activity.plot(kind='bar', color='lightcoral')
plt.title("User Activity - Frequency of Report Modifications", fontsize=16)
plt.xlabel("User", fontsize=12)
plt.ylabel("Number of Modifications", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyzing the number of reports generated per workspace
workspace_usage = df['workspace_name'].value_counts()

# Plotting the workspace usage
plt.figure(figsize=(12, 6))
workspace_usage.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.title("Workspace Usage - Number of Reports per Workspace", fontsize=16)
plt.xlabel("Workspace", fontsize=12)
plt.ylabel("Number of Reports", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Analyzing the number of reports stored on dedicated capacity vs general storage
storage_distribution = df['is_on_dedicated_capacity'].value_counts()

# Display the storage distribution
print("\nStorage Distribution (Dedicated Capacity vs General Storage):")
print(storage_distribution)

# Plotting the storage distribution
plt.figure(figsize=(8, 6))
storage_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'], startangle=90)

# Adding labels and title
plt.title("Storage Distribution - Dedicated Capacity vs General Storage", fontsize=16)

# Display the plot
plt.tight_layout()
plt.show()
```
